```python
# Crear una lista
team_list = list(resultado.itertuples(index=False, name=None))
# Crear el script subEvents
with open("SQL/team/team.sql", "w") as archivo:
    archivo.write("INSERT INTO teams(wyId, name, city, officialName) VALUES\n")

    valores_previos = ",\n".join(f"('{wyId}', '{name}', '{city}', '{officialName}')" for wyId, name, city, officialName in team_list)
    wyId = 1000  # Puedes cambiar el valor inicial seg√∫n necesites
    valores_nuevos = ",\n".join(f"('{i}', '{name}', 'London', 'N/S')" for i, name in enumerate(diferencia, start=wyId))
    valores_actualizados = valores_previos + ",\n" + valores_nuevos
    archivo.write(valores_actualizados+ ";\n")
```
